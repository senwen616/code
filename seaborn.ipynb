```python
# -*- coding: utf-8 -*-
%matplotlib inline
import numpy as np 
import seaborn as sns
import statsmodels
sns.set()
```

```python
x=np.random.randn(100)  #随机生成100个符合正态分布的数sns.kdeplot(x)
```

```python
sns.kdeplot(x)
```

```python
sns.kdeplot(x,cut=3)
```

```python
sns.kdeplot(x,cumulative=True)
```

```python
sns.kdeplot(x,shade=True)
```

```python
y = np.random.randn(100)
```

```python
sns.kdeplot(x,y,shade=True,color='b',cbar=True,cut=5)
```

```python
sns.distplot(x)
```

```python
import matplotlib.pyplot as plt
axes=plt.subplots(1,3)
sns.distplot(x,ax=axes[0])
sns.distplot(x,hist=False,ax=axes[1]) 
sns.distplot(x,kde=False,ax=axes[2]) 
```
